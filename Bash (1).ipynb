1. File Management script :-

   Write a Bash script that:-
      Creates a directory named "backup" in the user's home directory
      Copies all .txt files from the current directory into the "backup" directory
      Appends the current date and time to the filenames of the copied files.
  
      -
          #!/bin/bash

          BACKUP_DIR="$HOME/backup"  //dir path

          mkdir -p "$BACKUP_DIR" //create dir path

          TIMESTAMP=$(date +"%Y%m%d_%H%M%S") //current date & time

          for file in *.txt; do // search all .txt files

          [ -e "$file" ] || continue

          filename="${file%.*}"  // Extract filename without extension

          extension="${file##*.}"  // Get file extension

          //Copy file to backup directory with timestamp appended

          cp "$file" "$BACKUP_DIR/${filename}_${TIMESTAMP}.${extension}"
          done

          echo "Backup completed! All .txt files copied to $BACKUP_DIR with timestamp."

2.  System Health Check :-

     Create a script that:-
      Checks the systemâ€™s CPU and memory usage
      Reports if the CPU usage is above 80% or if the available memory is below 20%
      Logs the results to a file named system_health.log

      -
        #!/bin/bash

        LOG_FILE="system_health.log" //file name

        TIME=$(date) // get date

        CPU_USAGE=$(top -bn1 | grep "Cpu(s)" | awk '{print 100 - $8}') //CPU usage

        MEM_USAGE=$(free | awk '/Mem:/ {printf "%.0f", $3/$2 * 100}') //memory usage

        echo "------ System Check at $TIME ------" >> $LOG_FILE
        echo "CPU Usage: $CPU_USAGE%" >> $LOG_FILE
        echo "Memory Usage: $MEM_USAGE%" >> $LOG_FILE //write in log file

        if (( ${CPU_USAGE%.*} > 80 )); then
          echo "Warning: CPU usage above 80%!" >> $LOG_FILE
        fi //check cpu usage

      
        if (( MEM_USAGE > 80 )); then
          echo "Warning: Memory usage above 80%!" >> $LOG_FILE
        fi //check memory usage

        echo "" >> $LOG_FILE

4. Automated Backup

      Create a script that
      Takes a directory path as input from the user
      Compresses the directory into a .tar.gz file
      Saves the compressed file with a name that includes the current date (e.g., backup_2023-08-20.tar.gz)

    -  
        #!/bin/bash

        echo "Enter the directory path u want to backup:"

        read dir_path

        if [ ! -d "$dir_path" ]; then
              echo "the directory doesn't exist"
              exit 1
        fi

        curr_date=$(date +%Y-%m-%d)

        backup_file="backup_${curr_date}.tar.gz"

        tar -czf "$backup_file" "$dir_path"

        echo "backup created successfully"

        echo "file saved as:$backup_file"

5.  Simple To-Do List
    
     Implements a simple command-line to-do list
      Allows the user to add tasks, view tasks, and remove tasks
      Saves the tasks to a file (e.g., todo.txt).

       -  
            #!/bin/bash

            todo="todo.txt"

            if [ ! -f "$todo" ]; then
                    touch "$todo.txt"
            fi

            show_menu(){
              echo "-----Todo List App-----"

              echo "1. View Tasks"

              echo "2. Add Task"

              echo "3. Remove Task"

              echo "4. Exit"
            }

            while true
            do
                    show_menu

                    echo "Enter your choice:"

                    read choice


                    case $choice in
                            1)
                                    echo "Your tasks:"
                                    if [ -s "$todo" ]; then
                                            nl -w2 -s'. ' "$todo"
                                    else
                                            echo "No tasks yet"
                                    fi
                                    ;;

                            2)
                                    echo "Enter a new task:"
                                    read task
                                    echo "$task" >> "$todo"
                                    echo "Task Added"
                                    ;;

                            3)
                                    echo "Enter a task number to remove:"
                                    nl -w2 -s'. ' "$todo"
                                    read num
                                    if [ "$num" -le "$(wc -l < $todo)" ] && [ "$num" -gt 0 ]; then
                                            sed -i "${num}d" "$todo"
                                            echo "Task removed!"
                                    else
                                            echo "Invalid tasks"
                                    fi
                                    ;;

                            4)
                                    echo "Exiting the program...."
                                    break
                                    ;;

                            *)
                                    echo "Invalid choice!! Try again"
                                    ;;
                    esac
            done


6.  Automated Software Installation

      Reads a list of software package names from a file (e.g., packages.txt)
      Installs each package using the appropriate package manager (apt, yum, etc.)
      Logs the installation status of each package.

      -
          #!/bin/bash

          input="packages.txt"

          log_file="install_log.txt"

          if [ ! -f "log_file" ]; then
                  touch "$log_file"
          fi

          if [ ! -f "packages.txt" ]; then
                  echo "Packages file doesn't exists"
                  exit 1
          fi

          echo "starting the package installation"
          
          echo "Log file: $log_file"

          echo "---------------" >> "$log_file"

          echo "Installation started on : $(date)" >> "$log_file"

          echo "---------------" >> "$log_file"


          while read package; do
                  if [ -n "$package" ]; then
                          echo "Installing $package...."

                          if sudo apt install -y "$package" > /dev/null 2>&1; then
                                  echo "$package: Successfully done" >> "$log_file"
                          else
                                  echo "$package: Failed" >> "$log_file"
                          fi
                  fi
          done < "$input"

          echo "---------------" >> "$log_file"

          echo "Installation finished on: $(date)" >> "$log_file"

          echo "Check install_log.txt for results"

7. Text File Processing

   Takes a text file as input
   Counts and displays the number of lines, words, and characters in the file
   Finds and displays the longest word in the file.
   -
      #!/bin/bash

      read -p "Enter a filename:" filename

      if [ ! -f "$filename" ]; then
              echo "File doesn't exists"
              exit 1
      fi

      lines=$(wc -l < "$filename")

      words=$(wc -w < "$filename")

      chars=$(wc -m < "$filename")

      echo "File Data:"

      echo "\n"

      echo "Lines: $lines"

      echo "Words: $words"

      echo "Characters: $chars"

      long_w=""
      for word in $(cat "$filename"); do
              if [ ${#word} -gt ${#long_w} ]; then
                      long_w=$word
              fi
      done

      echo "Longest word is : $long_w"
